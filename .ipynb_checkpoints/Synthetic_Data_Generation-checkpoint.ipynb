```python
import pandas as pd
import numpy as np
import random
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
# Apply one-hot encoding to only 'City' and 'Road Type'
encoded_columns = pd.get_dummies(df[['City', 'Road_Type','Furnishing']], dtype=int)

# Drop the original 'City' and 'Road Type' columns
df = df.drop(['City', 'Road_Type','Furnishing'], axis=1)

# Concatenate the one-hot encoded columns to the DataFrame
df = pd.concat([df, encoded_columns], axis=1)
```

```python
# Step 1: compute frequency of each category
freq_map = df['Neighborhood'].value_counts(normalize=True)

# Step 2: map the frequencies to the column
df['Neighborhood_freq'] = df['Neighborhood'].map(freq_map)
```

```python
df = df.drop(['Neighborhood'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Price'] = scaler.fit_transform(df[['Price']])
```

```python
corr = df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(14, 6))

# Draw the heatmap
sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0,
    annot=False,  # well annotate selected below
    fmt=".2f",
    linewidths=.5,
    cbar_kws={"shrink": .5}
)

plt.title(" Correlation Matrix of All Features (Clean)", fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```

```python
# Target total rows
target_rows = 10_000
current_rows = len(df)
rows_to_generate = target_rows - current_rows

print(f"Current rows: {current_rows}")
print(f"Rows to generate: {rows_to_generate}")
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Function to generate synthetic data
def generate_synthetic_data(df, n):
    """
    Generate synthetic data by sampling rows and adding controlled noise.

    Parameters:
    df (pd.DataFrame): Original dataset
    n (int): Number of synthetic rows to generate

    Returns:
    pd.DataFrame: Synthetic dataset of n rows
    """
    synthetic_rows = []

    for i in range(n):
        # Step 1: Randomly sample an existing row
        real_row = df.sample(1, replace=True).iloc[0].copy()
        
        # Step 2: Modify numeric & discrete features
        for col in df.columns:
            if col in ['Bedroom', 'Bathroom', 'Amenities_Count']:
                # For integer count features: jitter by 1 (but 0)
                jitter = np.random.choice([-1, 0, 1])
                real_row[col] = max(0, real_row[col] + jitter)

            elif df[col].dtype in [np.float64, np.float32, np.int64, np.int32]:
                # For continuous features: add small Gaussian noise
                if col not in ['Bedroom', 'Bathroom', 'Amenities_Count'] and not col.startswith("City_") \
                   and not col.startswith("Road_Type_") and not col.startswith("Furnishing_"):
                    std_dev = df[col].std() * 0.05  # 5% of std
                    noise = np.random.normal(0, std_dev)
                    real_row[col] = real_row[col] + noise

            # One-hot encoded columns are left unchanged
        
        synthetic_rows.append(real_row)
    
    return pd.DataFrame(synthetic_rows)
```

```python
# Generate synthetic rows
synthetic_df = generate_synthetic_data(df, rows_to_generate)

# Combine original and synthetic data
augmented_df = pd.concat([df, synthetic_df], ignore_index=True)

print(f"Augmented dataset has {len(augmented_df)} rows.")
```

```python
augmented_df.to_csv("house_data.csv", index=False)
```

```python
corr = augmented_df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(14, 6))

# Draw the heatmap
sns.heatmap(
    corr,
    cmap="coolwarm",
    center=0,
    annot=False,  # well annotate selected below
    fmt=".2f",
    linewidths=.5,
    cbar_kws={"shrink": .5}
)

plt.title(" Correlation Matrix of All Features (Clean)", fontsize=16)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.show()
```
